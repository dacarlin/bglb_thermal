```python
import pandas 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pandas.read_csv( 'logo-data-bagel-aligned.txt', sep='\s+' ) 
df['pos'] = df.index + 1 
df['native_res'] = df.amino_acid.str.upper() 

df.columns
```

```python
AAs = [ u'A', u'C', u'D', u'E', u'F', u'G', u'H', u'I', u'K', u'L', u'M', u'N', u'P', u'Q', u'R', u'S', u'T', u'V', u'W', u'Y', ]

my_df = pandas.DataFrame( columns=AAs )

for idx, row in df.iterrows():
    for AA in AAs:
        my_row = row[ AAs ].astype( int ) 
        my_index = row.native_res + str( row.pos ) + AA
        my_df.loc[ my_index ] = my_row 
        
my_df
```

```python
for idx, row in my_df.iterrows(): 
    print row.sum() 
```

```python
plt.figure( figsize=( 30, 8 ) ) 
cols = []
for i in df.columns[2:22]:
    plt.plot( df.pos, df[i] )
    cols += [ i ] 
    
plt.legend( cols ) 
```
